```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Memuat citra dalam mode grayscale
image = cv2.imread('/content/image1.png', cv2.IMREAD_GRAYSCALE)

# Menampilkan citra asli
plt.subplot(131), plt.imshow(image, cmap='gray'), plt.title('Citra Asli')
```

```python

# Mendefinisikan kernel untuk operasi erosi dan dilasi
kernel = np.ones((5,5), np.uint8)
```

```python
#Melakukan operasi erosi
erosion = cv2.erode(image, kernel, iterations = 1)
plt.subplot(132), plt.imshow(erosion, cmap='gray'), plt.title('Erosi')

# Melakukan operasi dilasi
dilation = cv2.dilate(image, kernel, iterations = 1)
plt.subplot(133), plt.imshow(dilation, cmap='gray'), plt.title('Dilasi')

# Menampilkan hasil
plt.show()
```
